```python
query_lima_cat_a = """
SELECT DISTINCT 
    m.Ejecutora as Municipalidad, 
    m.Categoria_Municipal,
    g.Nombre_Provincia as Provincia,
    g.Nombre_Distrito as Distrito
FROM dim_municipalidad m
JOIN fact_ejec f 
  ON m.SK_Municipalidad = f.SK_Municipalidad
JOIN dim_geografia g 
  ON f.SK_Geografia = g.SK_Geografia
WHERE g.Nombre_Departamento = 'LIMA' 
  AND UPPER(m.Categoria_Municipal) LIKE '%A%'
ORDER BY Provincia, Distrito
"""

# Ejecutamos
spark.sql(query_lima_cat_a).toPandas()
```
