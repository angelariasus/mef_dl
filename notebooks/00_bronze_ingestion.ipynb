### Ejecutar Extracción Histórica (2012 a 2024)
Esta celda iterará desde 2012 hasta 2024, extrayendo toda la información histórica de la capa Bronze por cada año.

> **Nota:** El sistema es inteligente. Si una fuente ya fue descargada y no ha cambiado en origen, indicará `skipped` (omitido) y pasará a la siguiente sin descargar de nuevo.

```python
from core.config.settings import settings

print(f"Iniciando extracción de la capa Bronze ({min(settings.YEARS)} a {max(settings.YEARS)})...")
print("-" * 50)

# Extraer dinámicamente los años configurados
for year in settings.YEARS:
    print(f"\n--- Procesando año {year} ---")
    results = run_pipeline(year_filter=str(year))
    print(f"Exitosos: {results['successful']} | Omitidos: {results['skipped']} | Fallidos: {results['failed']}")

run_pipeline(source_prefix="sismepre")

print("\nCapa Bronze lista para ser procesada a Silver.")
```
