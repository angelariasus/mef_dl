```python
from bronze.ingest_bronze import run_pipeline
from src.paths import ensure_dirs

# Asegurarnos de que los directorios existen antes de extraer
ensure_dirs()
print("Módulos importados y directorios listos.")
```

### Ejecutar Extracción Completa
Esta celda ejecutará todo el pipeline para todas las fuentes configuradas en `bronze/config.py`.

> **Nota:** El sistema es inteligente. Si una fuente ya fue descargada y no ha cambiado en origen, indicará `skipped` (omitido) y pasará a la siguiente, ahorrando tiempo y ancho de banda.

```python
print("Iniciando extracción de la capa Bronze...")
print("-" * 50)

# Ejecutar sin filtros para descargar todo
results = run_pipeline()

print("\nResumen de la Ingesta:")
print("-" * 50)
print(f"Totales    : {results['total_sources']}")
print(f"Exitosos   : {results['successful']}")
print(f"Omitidos   : {results['skipped']} (ya estaban actualizados)")
print(f"Fallidos   : {results['failed']}")

if results['failed'] > 0:
    print("\nErrores encontrados:")
    for err in results['errors']:
        print(f"  - {err['name']}: {err['error']}")
else:
    print("\nCapa Bronze lista para ser procesada a Silver.")
```
